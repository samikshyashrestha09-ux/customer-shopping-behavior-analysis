````python
# Fill missing Review Ratings with the median rating of their own category
# This is smarter than using a global median since different categories may have different rating patterns
# lambda x represents each category group passed in one at a time by groupby
data['Review Rating'] = data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
````

````python
data.isnull().sum()
````

````python
# Convert column names to sanke case (lowercase and replace spaces with underscores)
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
#Renaming Column
data.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'}, inplace = True)
````

````python
data.columns
````

````python
# Divide age into 4 equal groups based on data distribution using quantile-based binning
# pd.qcut splits data into q=4 equal sized buckets (each containing 25% of the data)
# Labels assigned: Young Adult, Adult, Middle Aged, Senior
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
data['age_group'] = pd.qcut(data['age'], q=4, labels=labels)
````

````python
data[['age','age_group']].head(10)
````

````python
# Convert text-based purchase frequency into numeric days for easier analysis
# e.g. 'Weekly' → 7, 'Monthly' → 30, 'Annually' → 365
# .map() replaces each value in the column using the dictionary lookup
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

data['purchase_frequency_days'] = data['frequency_of_purchases'].map(frequency_mapping)
````

````python
data[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
````

````python
#Check if the values are actually same in discount_applied and promo_code_used
(data['discount_applied'] == data['promo_code_used']).all()
````

````python
# Dropping promo_code_used column
# axis = 1 means dropping column.
# axis = 0 means dropping rows
data = data.drop('promo_code_used', axis=1)
````

## Loading Cleaned Data into MySQL

After cleaning, the dataset was exported from the pandas DataFrame into a MySQL database for SQL-based analysis, using SQLAlchemy and PyMySQL as the connector.

**Approach:**
- Established a connection using `sqlalchemy.create_engine()` with a MySQL/PyMySQL connection string
- Wrote the cleaned DataFrame to a MySQL table using `DataFrame.to_sql()`, with `if_exists="replace"` to allow re-runs during development
- Verified the load with a `SELECT * FROM customer LIMIT 5` query read back via `pandas.read_sql()`

```python
from sqlalchemy import create_engine

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
data.to_sql("customer", engine, if_exists="replace", index=False)
```
